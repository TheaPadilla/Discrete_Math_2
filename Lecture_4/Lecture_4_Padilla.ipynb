**Kruskal's Algorithm**
<br> $\bullet$ It sorts the edges by weight and adds them one by one to create the minimum spanning tree, ensuring no cycles are formed. It's a greedy algorithm that works well for sparse graphs.
<br>**Prim's Algorithm**
<br> $\bullet$ It starts with a single node and grows the minimum spanning tree by adding the edge with the smallest weight that connects a visited node to an unvisited one. It's also a greedy algorithm, suitable for dense graphs.
<br>**Dijkstra's Shortest Path Algorithm**
<br>$\bullet$ It finds the shortest path between two nodes by exploring the graph, selecting the next node based on the smallest total weight from the start node. It's particularly useful for finding shortest paths in weighted graphs with non-negative weights.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the graph with weights
G = nx.Graph()
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G"])
edgelist = [
    ("A", "C", {'weight': 10}), ("A", "I", {'weight': 5}), ("C", "B", {'weight': 20}),
    ("C", "E", {'weight': 15}), ("C", "D", {'weight': 30}), ("B", "E", {'weight': 25}),
    ("I", "E", {'weight': 40}), ("E", "F", {'weight': 35}), ("F", "G", {'weight': 50}),
    ("G", "E", {'weight': 45})
]
G.add_edges_from(edgelist)

# Draw the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=55)
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A", "C"): 10, ("A", "I"): 5, ("C", "B"): 20, ("C", "E"): 15,
                 ("C", "D"): 30, ("B", "E"): 25, ("I", "E"): 40, ("E", "F"): 35,
                 ("F", "G"): 50, ("G", "E"): 45},
    font_color='red'
)
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    with_labels=True
)
plt.show()
```
